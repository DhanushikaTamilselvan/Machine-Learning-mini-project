```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = pd.read_csv('/content/Titanic-Dataset.csv')

# Select the features and target variable
X = df[['Pclass', 'Age', 'SibSp', 'Parch', 'Fare']]
y = df['Survived']  # 'Survived' is the target column for binary classification

# Convert the 'Sex' column to numeric using label encoding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])

# Remove rows with missing values
df_clean = df.dropna(subset=['Pclass', 'Age', 'SibSp', 'Parch', 'Fare', 'Survived'])
X_clean = df_clean[X.columns]
y_clean = df_clean[y.name]

# Split the cleaned data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_clean, y_clean, test_size=0.2, random_state=42)

# Create and train the Naive Bayes model
model = GaussianNB()
model.fit(X_train, y_train)

# Predict the classes
y_pred = model.predict(X_test)

# Calculate accuracy and print classification report
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print("Classification Report:\n", report)

# Create a confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Plot a confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Create a scatter plot for Age vs. Fare
plt.figure(figsize=(10, 6))
plt.scatter(df_clean['Age'], df_clean['Fare'], c=df_clean['Survived'], cmap='viridis')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot of Age vs. Fare (Survived: 1, Not Survived: 0)')
plt.show()
```
